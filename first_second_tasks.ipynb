```python
import pandas as pd
df_og = pd.read_csv("hackaton_mona_data/data/502acf0c3bfbe29dd8496a42634e85c7.csv")
df = df_og.copy()
```

```python
inches_to_mm = 25.4
df['rain_1h_mm'] = df['rain_1h']*inches_to_mm
df['rain_3h_mm'] = df['rain_3h']*inches_to_mm
df['rain_1h_mm'] = df['rain_1h_mm'].fillna(0)
df['rain_3h_mm'] = df['rain_1h_mm'].fillna(0)
average_rainfall_per_city = df[['city_name', 'rain_1h_mm', 'rain_3h_mm']].groupby(['city_name']).mean()
average_rainfall_per_city
```

```python
df['date_iso'] = pd.to_datetime(df['dt_iso'], format='%Y-%m-%d %H:%M:%S %z UTC')

df['date'] = df['date_iso'].dt.date
df = df.sort_values(by=['timezone', 'date_iso']).reset_index(drop=True)

total_rainfall_in_24H = df[['date', 'city_name', 'rain_1h_mm']].groupby(['city_name', 'date']).sum().reset_index()
total_rainfall_in_24H = total_rainfall_in_24H.rename(columns={'rain_1h_mm':'rain_24h_mm'})

df = df.merge(total_rainfall_in_24H, on=['city_name', 'date'], how='left')

flood_risk_threshold = 50
df['flood_risk'] = df['rain_24h_mm'] > flood_risk_threshold
df['flood_risk']
```
